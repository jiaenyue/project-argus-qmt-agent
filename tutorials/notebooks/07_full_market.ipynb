## 1. æ·±åº¦è¡Œæƒ…æ•°æ®å¤„ç†ç±»

```python
class MarketDepthAnalyzer:
    """æ·±åº¦è¡Œæƒ…æ•°æ®åˆ†æå™¨

    æä¾›å®Œæ•´è¡Œæƒ…æ•°æ®çš„å¤„ç†ã€åˆ†æå’Œå¯è§†åŒ–åŠŸèƒ½ã€‚
    æ”¯æŒå¤šç§æ•°æ®æºå’Œé«˜æ•ˆçš„å¤§æ•°æ®å¤„ç†ã€‚
    """

    def __init__(self, api_client: QMTAPIClient):
        """åˆå§‹åŒ–æ·±åº¦è¡Œæƒ…åˆ†æå™¨

        Args:
            api_client: APIå®¢æˆ·ç«¯å®ä¾‹
        """
        self.api_client = api_client
        self.data_cache = {}  # æ•°æ®ç¼“å­˜
        self.subscriptions = {}  # è®¢é˜…ç®¡ç†
        self.data_queue = queue.Queue()  # æ•°æ®å¤„ç†é˜Ÿåˆ—
        self.is_running = False
        self.processor_thread = None

        # æ€§èƒ½ç›‘æ§
        self.performance_monitor = PerformanceMonitor()

        # å¸‚åœºç»Ÿè®¡æ•°æ®
        self.market_stats = {
            "SH": {"up_count": 0, "down_count": 0, "flat_count": 0, "total": 0},
            "SZ": {"up_count": 0, "down_count": 0, "flat_count": 0, "total": 0},
        }

        # æ¿å—ç»Ÿè®¡æ•°æ®
        self.sector_stats = defaultdict(
            lambda: {"up_count": 0, "down_count": 0, "flat_count": 0, "total": 0}
        )

    def start_processing(self):
        """å¯åŠ¨æ•°æ®å¤„ç†çº¿ç¨‹"""
        if self.is_running:
            print("æ•°æ®å¤„ç†å™¨å·²åœ¨è¿è¡Œä¸­")
            return

        try:
            self.is_running = True
            self.processor_thread = threading.Thread(target=self._process_data_queue)
            self.processor_thread.daemon = True
            self.processor_thread.start()
            print("æ·±åº¦è¡Œæƒ…æ•°æ®å¤„ç†å™¨å·²å¯åŠ¨")
        except Exception as e:
            print(f"å¯åŠ¨æ•°æ®å¤„ç†å™¨å¤±è´¥: {e}")
            self.is_running = False

    def stop_processing(self):
        """åœæ­¢æ•°æ®å¤„ç†"""
        try:
            self.is_running = False
            if self.processor_thread:
                self.processor_thread.join(timeout=5)
            print("æ·±åº¦è¡Œæƒ…æ•°æ®å¤„ç†å™¨å·²åœæ­¢")
        except Exception as e:
            print(f"åœæ­¢æ•°æ®å¤„ç†å™¨æ—¶å‡ºé”™: {e}")

    def _process_data_queue(self):
        """å¤„ç†æ•°æ®é˜Ÿåˆ—ï¼ˆåå°çº¿ç¨‹ï¼‰"""
        while self.is_running:
            try:
                # ä»é˜Ÿåˆ—è·å–æ•°æ®ï¼Œè¶…æ—¶1ç§’
                item = self.data_queue.get(timeout=1)

                # å¤„ç†æ•°æ®
                self._process_market_data(item)

                self.data_queue.task_done()

            except queue.Empty:
                continue
            except Exception as e:
                print(f"æ•°æ®å¤„ç†é”™è¯¯: {e}")

    def _process_market_data(self, data_item: Dict):
        """å¤„ç†å¸‚åœºæ•°æ®

        Args:
            data_item: å¸‚åœºæ•°æ®é¡¹
        """
        data_type = data_item.get("type", "unknown")
        data = data_item.get("data", {})

        if data_type == "full_market":
            # å¤„ç†å®Œæ•´è¡Œæƒ…æ•°æ®
            self._update_market_stats(data)

            # è®°å½•æ€§èƒ½ç»Ÿè®¡
            self.performance_monitor.record_api_call(
                "process_full_market", 0.005, True  # å¤„ç†æ—¶é—´å¾ˆçŸ­
            )
        elif data_type == "tick_data":
            # å¤„ç†é€ç¬”æˆäº¤æ•°æ®
            symbol = data_item.get("symbol", "")
            if symbol:
                self._process_tick_data(symbol, data)

    def _update_market_stats(self, market_data: List[Dict]):
        """æ›´æ–°å¸‚åœºç»Ÿè®¡æ•°æ®

        Args:
            market_data: å¸‚åœºæ•°æ®åˆ—è¡¨
        """
        # é‡ç½®ç»Ÿè®¡æ•°æ®
        for market in self.market_stats:
            self.market_stats[market] = {
                "up_count": 0,
                "down_count": 0,
                "flat_count": 0,
                "total": 0,
            }

        self.sector_stats.clear()

        # ç»Ÿè®¡æ¶¨è·Œå®¶æ•°
        for stock_data in market_data:
            symbol = stock_data.get("symbol", "")
            if not symbol:
                continue

            # æå–å¸‚åœºå’Œæ¿å—ä¿¡æ¯
            market = symbol.split(".")[-1] if "." in symbol else ""
            sector = stock_data.get("sector", "unknown")

            # åˆ¤æ–­æ¶¨è·Œ
            change = stock_data.get("change", 0)

            # æ›´æ–°å¸‚åœºç»Ÿè®¡
            if market in self.market_stats:
                self.market_stats[market]["total"] += 1

                if change > 0:
                    self.market_stats[market]["up_count"] += 1
                elif change < 0:
                    self.market_stats[market]["down_count"] += 1
                else:
                    self.market_stats[market]["flat_count"] += 1

            # æ›´æ–°æ¿å—ç»Ÿè®¡
            self.sector_stats[sector]["total"] += 1

            if change > 0:
                self.sector_stats[sector]["up_count"] += 1
            elif change < 0:
                self.sector_stats[sector]["down_count"] += 1
            else:
                self.sector_stats[sector]["flat_count"] += 1

    def _process_tick_data(self, symbol: str, tick_data: Dict):
        """å¤„ç†é€ç¬”æˆäº¤æ•°æ®

        Args:
            symbol: è‚¡ç¥¨ä»£ç 
            tick_data: é€ç¬”æˆäº¤æ•°æ®
        """
        # ç¼“å­˜æ•°æ®
        if symbol not in self.data_cache:
            self.data_cache[symbol] = {"ticks": [], "last_update": time.time()}

        # æ·»åŠ æ–°æ•°æ®
        self.data_cache[symbol]["ticks"].append(tick_data)
        self.data_cache[symbol]["last_update"] = time.time()

        # é™åˆ¶ç¼“å­˜å¤§å°
        max_ticks = 1000  # æœ€å¤šä¿ç•™1000æ¡é€ç¬”æ•°æ®
        if len(self.data_cache[symbol]["ticks"]) > max_ticks:
            self.data_cache[symbol]["ticks"] = self.data_cache[symbol]["ticks"][-max_ticks:]

    def get_market_stats(self) -> Dict:
        """è·å–å¸‚åœºç»Ÿè®¡æ•°æ®

        Returns:
            Dict: å¸‚åœºç»Ÿè®¡æ•°æ®
        """
        return {
            "markets": self.market_stats,
            "sectors": dict(self.sector_stats),
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        }

    def get_market_breadth(self) -> Dict:
        """è·å–å¸‚åœºå®½åº¦æŒ‡æ ‡

        Returns:
            Dict: å¸‚åœºå®½åº¦æŒ‡æ ‡
        """
        if not self.market_stats:
            print("  æ— æ³•è®¡ç®—å¸‚åœºå®½åº¦æŒ‡æ ‡ï¼šå¸‚åœºç»Ÿè®¡æ•°æ®ä¸ºç©º")
            print("  è¯·ç¡®ä¿å·²è®¢é˜…å¹¶æ¥æ”¶åˆ°å¸‚åœºæ•°æ®")
            print("  å»ºè®®æ£€æŸ¥ï¼š")
            print("    - æ•°æ®è®¢é˜…æ˜¯å¦æˆåŠŸ")
            print("    - æ˜¯å¦æœ‰è¶³å¤Ÿçš„æ•°æ®ç§¯ç´¯")
            print("    - APIè¿æ¥çŠ¶æ€æ˜¯å¦æ­£å¸¸")
            return {}
        
        breadth = {}
        valid_markets = 0

        for market, stats in self.market_stats.items():
            try:
                if not isinstance(stats, dict):
                    print(f"  å¸‚åœºå®½åº¦è®¡ç®—è­¦å‘Šï¼š{market}å¸‚åœºæ•°æ®æ ¼å¼é”™è¯¯")
                    continue
                
                required_fields = ["total", "up_count", "down_count", "flat_count"]
                missing_fields = [field for field in required_fields if field not in stats]
                if missing_fields:
                    print(f"  å¸‚åœºå®½åº¦è®¡ç®—è­¦å‘Šï¼š{market}å¸‚åœºç¼ºå°‘å¿…è¦å­—æ®µ {missing_fields}")
                    continue
                
                if stats["total"] <= 0:
                    print(f"  å¸‚åœºå®½åº¦è®¡ç®—è­¦å‘Šï¼š{market}å¸‚åœºæ— æœ‰æ•ˆè‚¡ç¥¨æ•°æ®")
                    continue
                
                # æ•°æ®ä¸€è‡´æ€§æ£€æŸ¥
                calculated_total = stats["up_count"] + stats["down_count"] + stats["flat_count"]
                if abs(calculated_total - stats["total"]) > 0:
                    print(f"  å¸‚åœºå®½åº¦è®¡ç®—è­¦å‘Šï¼š{market}å¸‚åœºæ•°æ®ä¸ä¸€è‡´")
                    print(f"    æ€»æ•°: {stats['total']}, è®¡ç®—æ€»æ•°: {calculated_total}")
                
                advance_decline_ratio = stats["up_count"] / max(stats["down_count"], 1)
                breadth_index = (stats["up_count"] - stats["down_count"]) / stats["total"] * 100

                breadth[market] = {
                    "advance_decline_ratio": round(advance_decline_ratio, 2),
                    "breadth_index": round(breadth_index, 2),
                    "up_percent": round(stats["up_count"] / stats["total"] * 100, 2),
                    "down_percent": round(stats["down_count"] / stats["total"] * 100, 2),
                    "flat_percent": round(stats["flat_count"] / stats["total"] * 100, 2),
                }
                valid_markets += 1
                
            except Exception as e:
                print(f"  å¸‚åœºå®½åº¦è®¡ç®—å‡ºé”™ï¼ˆ{market}ï¼‰ï¼š{e}")
                print("  è¯·æ£€æŸ¥å¸‚åœºæ•°æ®çš„æ ¼å¼å’Œå®Œæ•´æ€§")
        
        if valid_markets == 0:
            print("  å¸‚åœºå®½åº¦æŒ‡æ ‡è®¡ç®—å¤±è´¥ï¼šæ²¡æœ‰æœ‰æ•ˆçš„å¸‚åœºæ•°æ®")
            print("  å»ºè®®æ£€æŸ¥æ•°æ®æ”¶é›†å’Œå¤„ç†æµç¨‹")
        
        return breadth

    def get_sector_performance(self, top_n: int = 5) -> Dict:
        """è·å–æ¿å—è¡¨ç°æ’å

        Args:
            top_n: è¿”å›å‰Nä¸ªæ¿å—

        Returns:
            Dict: æ¿å—è¡¨ç°æ•°æ®
        """
        if not self.sector_stats:
            print("  æ— æ³•è®¡ç®—æ¿å—è¡¨ç°ï¼šæ¿å—ç»Ÿè®¡æ•°æ®ä¸ºç©º")
            print("  è¯·ç¡®ä¿å·²è®¢é˜…å¹¶æ¥æ”¶åˆ°æ¿å—æ•°æ®")
            print("  å»ºè®®æ£€æŸ¥ï¼š")
            print("    - æ¿å—æ•°æ®è®¢é˜…æ˜¯å¦æˆåŠŸ")
            print("    - æ˜¯å¦æœ‰è¶³å¤Ÿçš„æ¿å—æ•°æ®ç§¯ç´¯")
            print("    - APIè¿æ¥çŠ¶æ€æ˜¯å¦æ­£å¸¸")
            return {
                "top_sectors": [],
                "bottom_sectors": [],
                "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "error": "æ¿å—æ•°æ®ä¸ºç©º"
            }
        
        if top_n <= 0:
            print("  æ¿å—è¡¨ç°è®¡ç®—è­¦å‘Šï¼štop_nå‚æ•°æ— æ•ˆ")
            print("  ä½¿ç”¨é»˜è®¤å€¼5")
            top_n = 5
        
        # è®¡ç®—æ¿å—å¼ºåº¦æŒ‡æ ‡
        sector_strength = []
        insufficient_data_sectors = []

        for sector, stats in self.sector_stats.items():
            try:
                if not isinstance(stats, dict):
                    print(f"  æ¿å—è¡¨ç°è®¡ç®—è­¦å‘Šï¼š{sector}æ¿å—æ•°æ®æ ¼å¼é”™è¯¯")
                    continue
                
                required_fields = ["total", "up_count", "down_count"]
                missing_fields = [field for field in required_fields if field not in stats]
                if missing_fields:
                    print(f"  æ¿å—è¡¨ç°è®¡ç®—è­¦å‘Šï¼š{sector}æ¿å—ç¼ºå°‘å¿…è¦å­—æ®µ {missing_fields}")
                    continue
                
                if stats["total"] < 3:  # é™ä½æœ€å°è‚¡ç¥¨æ•°é‡è¦æ±‚ï¼Œä½†ç»™å‡ºè­¦å‘Š
                    insufficient_data_sectors.append(f"{sector}({stats['total']}åª)")
                    continue
                elif stats["total"] < 5:
                    print(f"  æ¿å—è¡¨ç°è®¡ç®—æç¤ºï¼š{sector}æ¿å—è‚¡ç¥¨æ•°é‡è¾ƒå°‘ï¼ˆ{stats['total']}åªï¼‰")
                    print("    å»ºè®®è‡³å°‘åŒ…å«5åªè‚¡ç¥¨ä»¥è·å¾—æ›´å¯é çš„æ¿å—å¼ºåº¦æŒ‡æ ‡")
                
                # æ•°æ®ä¸€è‡´æ€§æ£€æŸ¥
                if stats["up_count"] + stats["down_count"] > stats["total"]:
                    print(f"  æ¿å—è¡¨ç°è®¡ç®—è­¦å‘Šï¼š{sector}æ¿å—æ•°æ®ä¸ä¸€è‡´")
                    print(f"    ä¸Šæ¶¨+ä¸‹è·Œæ•°é‡({stats['up_count']+stats['down_count']}) > æ€»æ•°({stats['total']})")
                
                strength = (stats["up_count"] - stats["down_count"]) / stats["total"] * 100
                sector_strength.append(
                    {
                        "sector": sector,
                        "strength": round(strength, 2),
                        "up_count": stats["up_count"],
                        "down_count": stats["down_count"],
                        "total": stats["total"],
                        "up_percent": round(stats["up_count"] / stats["total"] * 100, 2),
                    }
                )
                
            except Exception as e:
                print(f"  æ¿å—è¡¨ç°è®¡ç®—å‡ºé”™ï¼ˆ{sector}ï¼‰ï¼š{e}")
                print("  è¯·æ£€æŸ¥æ¿å—æ•°æ®çš„æ ¼å¼å’Œå®Œæ•´æ€§")
        
        if insufficient_data_sectors:
            print(f"  æ¿å—è¡¨ç°è®¡ç®—æç¤ºï¼šä»¥ä¸‹æ¿å—å› è‚¡ç¥¨æ•°é‡ä¸è¶³è¢«è·³è¿‡ï¼š")
            for sector_info in insufficient_data_sectors[:5]:  # åªæ˜¾ç¤ºå‰5ä¸ª
                print(f"    {sector_info}")
            if len(insufficient_data_sectors) > 5:
                print(f"    ... è¿˜æœ‰{len(insufficient_data_sectors)-5}ä¸ªæ¿å—")
        
        if not sector_strength:
            print("  æ¿å—è¡¨ç°è®¡ç®—å¤±è´¥ï¼šæ²¡æœ‰æœ‰æ•ˆçš„æ¿å—æ•°æ®")
            print("  å»ºè®®æ£€æŸ¥ï¼š")
            print("    - æ¿å—æ•°æ®æ”¶é›†æ˜¯å¦æ­£å¸¸")
            print("    - æ¿å—å†…æ˜¯å¦åŒ…å«è¶³å¤Ÿçš„æ´»è·ƒè‚¡ç¥¨")
            print("    - æ•°æ®å¤„ç†æµç¨‹æ˜¯å¦æ­£ç¡®")
            return {
                "top_sectors": [],
                "bottom_sectors": [],
                "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "error": "æ²¡æœ‰æœ‰æ•ˆçš„æ¿å—æ•°æ®"
            }

        # æŒ‰å¼ºåº¦æ’åº
        sector_strength.sort(key=lambda x: x["strength"], reverse=True)
        
        # ç¡®ä¿ä¸è¶…è¿‡å¯ç”¨æ¿å—æ•°é‡
        actual_top_n = min(top_n, len(sector_strength))
        if actual_top_n < top_n:
            print(f"  æ¿å—è¡¨ç°è®¡ç®—æç¤ºï¼šå¯ç”¨æ¿å—æ•°é‡({len(sector_strength)})å°‘äºè¯·æ±‚æ•°é‡({top_n})")

        return {
            "top_sectors": sector_strength[:actual_top_n],
            "bottom_sectors": sector_strength[-actual_top_n:] if len(sector_strength) >= actual_top_n else [],
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        }

    def analyze_tick_data(self, symbol: str) -> Dict:
        """åˆ†æé€ç¬”æˆäº¤æ•°æ®

        Args:
            symbol: è‚¡ç¥¨ä»£ç 

        Returns:
            Dict: åˆ†æç»“æœ
        """
        if not symbol:
            print("  é€ç¬”æ•°æ®åˆ†æå¤±è´¥ï¼šè‚¡ç¥¨ä»£ç ä¸ºç©º")
            return {"error": "è‚¡ç¥¨ä»£ç ä¸ºç©º"}
        
        if symbol not in self.data_cache:
            print(f"  é€ç¬”æ•°æ®åˆ†æå¤±è´¥ï¼šæœªæ‰¾åˆ°è‚¡ç¥¨{symbol}çš„ç¼“å­˜æ•°æ®")
            print("  è¯·ç¡®ä¿å·²è®¢é˜…è¯¥è‚¡ç¥¨çš„é€ç¬”æ•°æ®")
            return {"error": "æ— ç¼“å­˜æ•°æ®", "symbol": symbol}
        
        if not self.data_cache[symbol]["ticks"]:
            print(f"  é€ç¬”æ•°æ®åˆ†æå¤±è´¥ï¼šè‚¡ç¥¨{symbol}çš„é€ç¬”æ•°æ®ä¸ºç©º")
            print("  å¯èƒ½çš„åŸå› ï¼š")
            print("    - æ•°æ®è®¢é˜…æ—¶é—´è¿‡çŸ­ï¼Œå°šæœªç§¯ç´¯è¶³å¤Ÿæ•°æ®")
            print("    - è¯¥è‚¡ç¥¨å½“å‰æ— äº¤æ˜“æ´»åŠ¨")
            print("    - æ•°æ®æ¥æ”¶è¿‡ç¨‹ä¸­å‡ºç°é—®é¢˜")
            return {"error": "æ— å¯ç”¨æ•°æ®", "symbol": symbol}

        ticks = self.data_cache[symbol]["ticks"]
        
        if len(ticks) < 10:
            print(f"  é€ç¬”æ•°æ®åˆ†æè­¦å‘Šï¼šè‚¡ç¥¨{symbol}çš„æ•°æ®é‡è¾ƒå°‘ï¼ˆ{len(ticks)}æ¡ï¼‰")
            print("  åˆ†æç»“æœå¯èƒ½ä¸å¤Ÿå‡†ç¡®ï¼Œå»ºè®®ç§¯ç´¯æ›´å¤šæ•°æ®")

        try:
            # æå–ä»·æ ¼å’Œæˆäº¤é‡ï¼Œå¹¶éªŒè¯æ•°æ®æœ‰æ•ˆæ€§
            prices = []
            volumes = []
            invalid_ticks = 0
            
            for i, tick in enumerate(ticks):
                price = tick.get("price", 0)
                volume = tick.get("volume", 0)
                
                if not isinstance(price, (int, float)) or price <= 0:
                    invalid_ticks += 1
                    continue
                if not isinstance(volume, (int, float)) or volume < 0:
                    invalid_ticks += 1
                    continue
                    
                prices.append(price)
                volumes.append(volume)
            
            if invalid_ticks > 0:
                print(f"  é€ç¬”æ•°æ®åˆ†æè­¦å‘Šï¼šå‘ç°{invalid_ticks}æ¡æ— æ•ˆæ•°æ®")
                print("  å·²è‡ªåŠ¨è¿‡æ»¤æ— æ•ˆæ•°æ®")
            
            if not prices:
                print(f"  é€ç¬”æ•°æ®åˆ†æå¤±è´¥ï¼šè‚¡ç¥¨{symbol}æ²¡æœ‰æœ‰æ•ˆçš„ä»·æ ¼æ•°æ®")
                return {"error": "æ— æœ‰æ•ˆä»·æ ¼æ•°æ®", "symbol": symbol}

            # è®¡ç®—ç»Ÿè®¡æŒ‡æ ‡
            avg_price = sum(prices) / len(prices)
            avg_volume = sum(volumes) / len(volumes) if volumes else 0
            price_volatility = np.std(prices) if len(prices) > 1 else 0

            # åˆ†æä¹°å–ç›˜å£
            buy_orders = [tick for tick in ticks if tick.get("direction") == "B"]
            sell_orders = [tick for tick in ticks if tick.get("direction") == "S"]
            neutral_orders = [tick for tick in ticks if tick.get("direction") not in ["B", "S"]]

            if neutral_orders:
                print(f"  é€ç¬”æ•°æ®åˆ†ææç¤ºï¼šå‘ç°{len(neutral_orders)}æ¡æ–¹å‘ä¸æ˜çš„äº¤æ˜“æ•°æ®")

            buy_volume = sum(tick.get("volume", 0) for tick in buy_orders if isinstance(tick.get("volume"), (int, float)))
            sell_volume = sum(tick.get("volume", 0) for tick in sell_orders if isinstance(tick.get("volume"), (int, float)))

            # è®¡ç®—ä¹°å–æ¯”
            if sell_volume > 0:
                buy_sell_ratio = buy_volume / sell_volume
            elif buy_volume > 0:
                buy_sell_ratio = float("inf")
                print("  é€ç¬”æ•°æ®åˆ†ææç¤ºï¼šæ— å–ç›˜æˆäº¤ï¼Œä¹°å–æ¯”ä¸ºæ— ç©·å¤§")
            else:
                buy_sell_ratio = 0
                print("  é€ç¬”æ•°æ®åˆ†ææç¤ºï¼šæ— æ˜ç¡®ä¹°å–æ–¹å‘çš„æˆäº¤æ•°æ®")

            return {
                "symbol": symbol,
                "tick_count": len(ticks),
                "valid_tick_count": len(prices),
                "avg_price": round(avg_price, 2),
                "avg_volume": round(avg_volume, 2),
                "price_volatility": round(price_volatility, 4),
                "buy_count": len(buy_orders),
                "sell_count": len(sell_orders),
                "buy_volume": buy_volume,
                "sell_volume": sell_volume,
                "buy_sell_ratio": round(buy_sell_ratio, 2) if buy_sell_ratio != float("inf") else "âˆ",
                "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            }
            
        except Exception as e:
            print(f"  é€ç¬”æ•°æ®åˆ†æå‡ºé”™ï¼ˆ{symbol}ï¼‰ï¼š{e}")
            print("  è¯·æ£€æŸ¥æ•°æ®æ ¼å¼å’Œè®¡ç®—é€»è¾‘")
            return {"error": f"åˆ†æå‡ºé”™: {e}", "symbol": symbol}

    def visualize_market_breadth(self):
        """å¯è§†åŒ–å¸‚åœºå®½åº¦æŒ‡æ ‡"""
        try:
            breadth = self.get_market_breadth()

            if not breadth:
                print("æš‚æ— å¸‚åœºå®½åº¦æ•°æ®å¯è§†åŒ–")
                return

            # åˆ›å»ºå›¾è¡¨
            fig, ax = plt.subplots(figsize=(10, 6))

            # å‡†å¤‡æ•°æ®
            markets = list(breadth.keys())
            up_percents = [breadth[m]["up_percent"] for m in markets]
            down_percents = [breadth[m]["down_percent"] for m in markets]
            flat_percents = [breadth[m]["flat_percent"] for m in markets]

            # ç»˜åˆ¶å †å æŸ±çŠ¶å›¾
            width = 0.35
            ax.bar(markets, up_percents, width, label="ä¸Šæ¶¨", color="red")
            ax.bar(markets, down_percents, width, bottom=up_percents, label="ä¸‹è·Œ", color="green")
            ax.bar(
                markets,
                flat_percents,
                width,
                bottom=[up + down for up, down in zip(up_percents, down_percents)],
                label="å¹³ç›˜",
                color="gray",
            )

            # æ·»åŠ æ ‡ç­¾å’Œå›¾ä¾‹
            ax.set_ylabel("ç™¾åˆ†æ¯”")
            ax.set_title("å¸‚åœºæ¶¨è·Œåˆ†å¸ƒ")
            ax.legend()

            # æ˜¾ç¤ºå›¾è¡¨
            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"å¯è§†åŒ–å¸‚åœºå®½åº¦æ—¶å‡ºé”™: {e}")

    def visualize_sector_performance(self, top_n: int = 5):
        """å¯è§†åŒ–æ¿å—è¡¨ç°

        Args:
            top_n: æ˜¾ç¤ºå‰Nä¸ªæ¿å—
        """
        try:
            sector_data = self.get_sector_performance(top_n)

            if not sector_data or not sector_data["top_sectors"]:
                print("æš‚æ— æ¿å—è¡¨ç°æ•°æ®å¯è§†åŒ–")
                return

            # åˆ›å»ºå›¾è¡¨
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

            # å‡†å¤‡æ•°æ® - æœ€å¼ºæ¿å—
            top_sectors = sector_data["top_sectors"]
            top_names = [s["sector"] for s in top_sectors]
            top_strengths = [s["strength"] for s in top_sectors]

            # ç»˜åˆ¶æœ€å¼ºæ¿å—
            bars1 = ax1.bar(top_names, top_strengths, color="red")
            ax1.set_title("æœ€å¼ºæ¿å—")
            ax1.set_ylabel("å¼ºåº¦æŒ‡æ ‡")

            # ä¸ºæŸ±çŠ¶å›¾æ·»åŠ æ•°å€¼æ ‡ç­¾
            for bar in bars1:
                height = bar.get_height()
                ax1.annotate(
                    f"{height:.1f}",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3ç‚¹å‚ç›´åç§»
                    textcoords="offset points",
                    ha="center",
                    va="bottom",
                )

            # å‡†å¤‡æ•°æ® - æœ€å¼±æ¿å—
            bottom_sectors = sector_data["bottom_sectors"]
            bottom_names = [s["sector"] for s in bottom_sectors]
            bottom_strengths = [s["strength"] for s in bottom_sectors]

            # ç»˜åˆ¶æœ€å¼±æ¿å—
            bars2 = ax2.bar(bottom_names, bottom_strengths, color="green")
            ax2.set_title("æœ€å¼±æ¿å—")
            ax2.set_ylabel("å¼ºåº¦æŒ‡æ ‡")

            # ä¸ºæŸ±çŠ¶å›¾æ·»åŠ æ•°å€¼æ ‡ç­¾
            for bar in bars2:
                height = bar.get_height()
                ax2.annotate(
                    f"{height:.1f}",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3 if height >= 0 else -15),
                    textcoords="offset points",
                    ha="center",
                    va="bottom",
                )

            # æ˜¾ç¤ºå›¾è¡¨
            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"å¯è§†åŒ–æ¿å—è¡¨ç°æ—¶å‡ºé”™: {e}")

    def get_performance_stats(self) -> Dict:
        """è·å–æ€§èƒ½ç»Ÿè®¡

        Returns:
            Dict: æ€§èƒ½ç»Ÿè®¡æ•°æ®
        """
        return self.performance_monitor.export_stats()
```
